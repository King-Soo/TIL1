```python
starbucksMap = folium.Map([37.57,126.98], zoom_start=15, tiles='Stamen Terrain')
for i in seoulStarbucks.index:
    lat = seoulStarbucks.loc[i,'위도'] # 한줄씩 읽어 출력
    lng = seoulStarbucks.loc[i,'경도']
    storeType = seoulStarbucks.loc[i,'매장타입']
    #데이터프레임의 각 행 또는 열 단위로 추출한 데이터를 시리즈(Series)라고 함
    if storeType == 'general':
        folium.CircleMarker(
        location = [lat, lng],
        radius=2,
        fill=True,
        fill_color='red',
        fill_opacity=1,
        color='gray'
        ).add_to(starbucksMap)
    elif storeType == 'reserve':
         folium.CircleMarker(
         location = [lat, lng],
         radius=2,
         fill=True,
         fill_color='red',
         fill_opacity=1,
         color='blue'
         ).add_to(starbucksMap)
    else:
         folium.CircleMarker(
         location = [lat, lng],
         radius=2,
         fill=True,
         fill_color='red',
         fill_opacity=1,
         color='red'
         ).add_to(starbucksMap)
```
