```python
pieces=[]

for year in range(1880,2011):
    df=pd.read_csv('yob/yob{0}.txt'.format(year),
                    header=None,
                    names=['name','sex','births'])
    df['year']=year
    pieces.append(df)
len(pieces)
```

```python
names=pd.concat(pieces, ignore_index=1)
names
```

```python
#feature enginneering(특성 공학): 데이터의 기존 변수로부터 연산을 통해 새로운 변수를 생성하는 작업
# ex) 국어, 영어 => 평균
```

```python
# 성별에 따른 births의 합계
# 1880~2010까지 태어난 아이들의 성별에 따른 수
names.groupby('sex').births.sum()

#names.groupby('sex').groupby('year)'.births.sum() #에러

names.groupby(['sex','year']).births.sum()
# 성별에 따른 년도별 births 합계
# 1880~2010년까지 태어난 아이들의 수를 연도로 나누어 성별에 따라 비교
```

```python
#pivot_table함수를 이용한 연산
total_births=names.pivot_table('births',index='year',columns='sex', aggfunc=max)
```

```python
def add_prop(gr):
    gr['prop']=gr.births/gr.births.sum()
    return gr




names=names.groupby(['year','sex']).apply(add_prop)
names
```

```python
def getTop1000(gr):
    return gr.sort_values('births',ascending=False)[:1000]
    
top1000=names.groupby(['year','sex']).apply(getTop1000)
top1000
```
