```python
fp=open("fruits-vegetables.html",encoding='utf-8')
soup=BeautifulSoup(fp,'html.parser')
soup
```

```python
print(soup.select("ul")[1])
print(soup.select("#ve-list"))
print("\n\n",soup.select("#ve-list > li[data-lo='us']"))
print(soup.select("#ve-list > li.black"))
print(soup.select("li.black"))
```

```python
print(soup.select("html > body > div > ul > li")[7].string)
print(soup.select("ul > li")[7].string)
print(soup.select("ul#ve-list > li")[3].string)
print(soup.select("ul#ve-list > li.black")[1].string)
print(soup.select("div#main-goods > ul#ve-list > li.black")[1].string)
print(soup.select("li.black")[1].string)

avo=soup.select("li.black")
for i in avo :
    if i.attrs['data-lo']=="us" :
        print(i.string)

```
