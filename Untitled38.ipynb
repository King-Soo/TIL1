```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

```python
pd.cut(ages,[0,20,30,50,70,100],labels=['미성년자','청년','중년','장년','노년'])
```

```python
bins=[0,20,30,50,70,100]
labels=['미성년자','청년','중년','장년','노년']
mycut=pd.cut(ages,
       bins,
      labels=labels)
```

```python
type(mycut)
```

```python
mycut.categories
```

```python
mycut.codes
```

```python
agedf=pd.DataFrame(ages, columns=['ages'])
agedf
```

```python
agedf['age_cut']=mycut
```
