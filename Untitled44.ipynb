```python
# merge함수 : 공통 열 or 행 기준으로 두 데이터프레임을 합침
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
```

```python
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
```

```python
df2
```

```python
df1.merge(df2,left_on='고객번호',right_on='고객번호')
```

```python
pd.merge(df1,df2)
```

```python
df1.merge(df2)
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1,df2)
```

```python
df12=pd.merge(df1,df2, on='고객명')
df12
```

```python
데이터x => 금액, 데이터y => 성별 열이름으로 변경
```

```python
df12.rename(columns={
    '데이터_x':'금액',
    '데이터_y':'성별'}, inplace=1)
df12
```
